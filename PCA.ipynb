```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df
```

```python
dataset.target
```

```python
df.describe()
```

```python
X = df
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
df_scaled = pd.DataFrame(data = X_scaled, columns=dataset.feature_names)
df_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

<h3>Use PCA to reduce dimensions</h3>

<h4>Use components such that 95% of variance is retained using scaled data</h4>

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca_Scaled = pca.fit_transform(X_scaled)
X_pca_Scaled.shape
```

```python
X_pca_Scaled[0]
```

```python
df_afterpca = pd.DataFrame(data = X_pca_Scaled)
df_afterpca
```

```python
X_train_pca_Scaled, X_test_pca_Scaled, y_train, y_test = train_test_split(X_pca_Scaled, y, test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca_Scaled, y_train)
model.score(X_test_pca_Scaled, y_test)
```

**Let's now select only two components**

```python
pca = PCA(n_components=20)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
